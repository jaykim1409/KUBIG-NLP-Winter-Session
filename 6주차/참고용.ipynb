```python
import tensorflow as tf
from transformers import TFBertForNextSentencePrediction
from transformers import AutoTokenizer
```

```python
model = TFBertForNextSentencePrediction.from_pretrained('klue/bert-base', from_pt=True)
tokenizer = AutoTokenizer.from_pretrained("klue/bert-base")
```

```python
# 이어지는 두 개의 문장
prompt = "일부 학계에서는 챗GPT를 학생들이 표절에 사용할 수 있다며 우려를 표했다."
next_sentence = "실제로 챗GPT는 의사면허시험 합격 성적을 받기도 했고 경영학 석사(MBA) 시험도 통과했다. \
                이에 프린스턴 대학에 재학 중인 한 학생이 AI가 사용됐는지 식별해내는 'GPT제로' 프로그램을 개발하기도 했다."
encoding = tokenizer(prompt, next_sentence, return_tensors='tf')

logits = model(encoding['input_ids'], token_type_ids=encoding['token_type_ids'])[0]

softmax = tf.keras.layers.Softmax()
probs = softmax(logits)
print('최종 예측 레이블 :', tf.math.argmax(probs, axis=-1).numpy())
```

```python
# 상관없는 두 개의 문장
prompt = "일부 학계에서는 챗GPT를 학생들이 표절에 사용할 수 있다며 우려를 표했다."
next_sentence = "영서와 충북 북부에 1cm 미만으로 눈의 양이 많지는 않겠는데요, \
           중서부 지역을 중심으로도 약한 눈이 날릴 수 있기 때문에 교통안전에는 유의하셔야겠습니다."
encoding = tokenizer(prompt, next_sentence, return_tensors='tf')

logits = model(encoding['input_ids'], token_type_ids=encoding['token_type_ids'])[0]

softmax = tf.keras.layers.Softmax()
probs = softmax(logits)
print('최종 예측 레이블 :', tf.math.argmax(probs, axis=-1).numpy())
```

```python
import pandas as pd
from transformers import BertTokenizer

tokenizer = BertTokenizer.from_pretrained("bert-base-uncased") # Bert-base의 토크나이저
```

```python
result = tokenizer.tokenize("You're my chemical hype boy")
print(result)
```

```python
result = tokenizer.tokenize("Tastes like strawberries on a summer evening")
print(result)
```
